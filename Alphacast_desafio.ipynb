2. **Tiempo de Espera (Timeout)**:
   - Se añadió un parámetro `timeout` a la llamada `requests.get()` para evitar que el script se quede colgado indefinidamente.

## Crear dataset random para visualizar objetivo

````python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Define the base data
regions = ['GBA', 'Pampeana', 'Noroeste', 'Noreste', 'Cuyo', 'Patagonia']
products = [
    ('Pan francés', 'kg'),
    ('Fideos secos tipo guisero', '500 g'),
    ('Aceite de girasol', '1,5 litros'),
    ('Arroz blanco', 'kg'),
    ('Harina de trigo', 'kg'),
    ('Leche entera', '1 litro'),
    ('Papas', 'kg'),
    ('Tomates', 'kg')
]

# Generate dates for 1 year of data
start_date = datetime(2017, 6, 1)
dates = [start_date + timedelta(days=x) for x in range(0, 365)]

# Create empty lists to store data
data = []

# Generate data
for date in dates:
    for region in regions:
        for product, unit in products:
            # Generate price with some randomization and regional variation
            base_price = {
                'Pan francés': 35.0,
                'Fideos secos tipo guisero': 18.0,
                'Aceite de girasol': 51.0,
                'Arroz blanco': 25.0,
                'Harina de trigo': 30.0,
                'Leche entera': 22.0,
                'Papas': 20.0,
                'Tomates': 40.0
            }[product]

            # Add regional variation
            region_factor = {
                'GBA': 1.1,
                'Pampeana': 1.0,
                'Noroeste': 1.15,
                'Noreste': 1.2,
                'Cuyo': 1.05,
                'Patagonia': 1.3
            }[region]

            # Add random variation and time-based trend
            random_factor = np.random.normal(1, 0.05)
            time_factor = 1 + (dates.index(date) / 365) * 0.15  # 15% increase over the year

            price = base_price * region_factor * random_factor * time_factor

            data.append({
                'Date': date.strftime('%Y-%m-%d'),
                'Region': region,
                'Product': product,
                'Unit': unit,
                'Price': round(price, 2)
            })

# Create DataFrame
df = pd.DataFrame(data)

# Sort by date, region, and product
df = df.sort_values(['Date', 'Region', 'Product'])

# Display first few rows
df.head(10)

# Optional: Save to CSV
df.to_csv('price_data_2017.csv', index=False)
````

## Uso de RegEx para limpieza de columnas

Podemos usar expresiones regulares para  renombrar cualquier columna que contenga 'Año'
y dejando solo el valor numérico, por ejemplo, 2017.


````python
import pandas as pd
import re

def rename_year_columns(df):
    """
    Renombra columnas que contenga 'Año' a solo valor numerico.
    """
    new_columns = []
    for col in df.columns:
        match = re.search(r'Año\s*(\d{4})', col)  # Usamos regex para encontrar 'Año'
        if match:
            new_columns.append(match.group(1))    # Nos quedamos solo con el numero
        else:
            new_columns.append(col)
    df.columns = new_columns
    return df

df = rename_year_columns(df)
print(df.columns) # Print the updated column names
````

### Script Mejorado

> Idea: En el loop del script agregar codigo para llenar campos con la fecha standard yyyy-mm-dd, comenzando desde 2017-06-01. Luego borrar la segunda fila que contiene Mes .





````python
# Script mejorado

import pandas as pd

# Create a mapping for renaming columns
column_mapping = {}

# Initialize starting year and month
current_year = 2017
current_month = 6

# Loop through the column names
for column in df.columns:
    if 'Unnamed' in column or column.isdigit():  # Check if the column is unnamed or a year
        # Create the new column name in the format "YYYY-MM-DD"
        new_column_name = f"{current_year}-{current_month:02d}-01"
        column_mapping[column] = new_column_name

        # Increment the month and year
        current_month += 1
        if current_month > 12:  # If the month exceeds December, reset to January and increment the year
            current_month = 1
            current_year += 1

# Rename the columns using the mapping
df2 = df.rename(columns=column_mapping)

# Print updated column names
print("\nUpdated column names:")
print(df2.columns)

# Print the updated DataFrame
print("\nUpdated DataFrame:")
print(df2.head())
````
